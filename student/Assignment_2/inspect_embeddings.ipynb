## Sentence Similarity

```python
!pip -q install nltk scikit-learn umap-learn
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize

def tokenize_text(text):
    return word_tokenize(text.lower())
```

```python
# List of sentences for comparison
sentences = [
    "The cat sat on the mat.",
    "A dog is playing in the yard.",
    "I need to buy groceries today.",
    "The feline was resting on the rug.",
    "Canines enjoy outdoor activities.",
    "My shopping list includes milk and bread.",
    "The weather is beautiful today.",
    "Programming requires logical thinking."
]

# Query sentence
query = "My pet cat is sleeping on the carpet."
```

```python
import pickle
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

with open("word2vec_embeddings.pkl", "rb") as f:
    pt = pickle.load(f)

pt_embeddings = np.asarray(pt["embeddings"])     # shape: [V, D]
pt_word2idx = pt["word2idx"]

with open("word2vec_gensim_embeddings.pkl", "rb") as f:
    gs = pickle.load(f)

gs_embeddings = np.asarray(gs["embeddings"])     # shape: [V, D]
gs_word2idx = gs["word2idx"]

print("PyTorch embeddings shape:", pt_embeddings.shape)
print("Gensim embeddings shape:", gs_embeddings.shape)
```

```python
def get_word_vec_matrix(word, embeddings, word2idx):
    idx = word2idx.get(word)
    if idx is None:
        return None
    return embeddings[idx]

def sentence_embedding(text, embeddings, word2idx):
    tokens = tokenize_text(text)
    vecs = []
    for t in tokens:
        v = get_word_vec_matrix(t, embeddings, word2idx)
        if v is not None:
            vecs.append(v)
    if len(vecs) == 0:
        # no known words → return zero vector
        return np.zeros(embeddings.shape[1], dtype=float)
    return np.mean(np.vstack(vecs), axis=0)

def print_sentence_similarities(query, sentences, embeddings, word2idx, label):
    q = sentence_embedding(query, embeddings, word2idx).reshape(1, -1)
    print(f"\n=== {label} ===")
    for s in sentences:
        v = sentence_embedding(s, embeddings, word2idx).reshape(1, -1)
        sim = float(cosine_similarity(q, v)[0, 0])
        print(f"{sim: .4f}  | {s}")
```

```python
print_sentence_similarities(query, sentences, pt_embeddings, pt_word2idx, "PyTorch Word2Vec")
print_sentence_similarities(query, sentences, gs_embeddings, gs_word2idx, "Gensim-trained Word2Vec")
```

## Embedding Visualization

```python
animals = [
    'dog', 'cat', 'horse', 'cow',
    'lion', 'tiger', 'elephant', 'giraffe',
    'eagle', 'hawk', 'sparrow', 'penguin',
    'shark', 'whale', 'dolphin', 'tuna'
]

foods = [
    'apple', 'orange', 'banana', 'grape',
    'bread', 'pasta', 'rice', 'potato',
    'chicken', 'beef', 'pork', 'fish',
    'coffee', 'tea', 'juice', 'water'
]
```

```python
def collect_vectors_matrix(words, embeddings, word2idx, label):
    X, y, labels = [], [], []
    for w in words:
        if w in word2idx:
            X.append(embeddings[word2idx[w]])
            y.append(label)
            labels.append(w)
    return X, y, labels

def collect_vectors_kv(words, kv, label):
    X, y, labels = [], [], []
    for w in words:
        if w in kv:
            X.append(kv[w])
            y.append(label)
            labels.append(w)
    return X, y, labels
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import umap

def plot_2d(X, y, labels, title):
    plt.figure(figsize=(7, 6))
    X = np.array(X)
    y = np.array(y)

    plt.scatter(X[y==0, 0], X[y==0, 1], c="tab:blue", label="Animals")
    plt.scatter(X[y==1, 0], X[y==1, 1], c="tab:orange", label="Foods")

    for i, word in enumerate(labels):
        plt.annotate(word, (X[i, 0], X[i, 1]), fontsize=9)

    plt.legend()
    plt.title(title)
    plt.show()


def run_tsne(X):
    X = np.array(X)  # <-- FIX
    n = X.shape[0]
    perplexity = min(10, max(2, n // 3))  # safe heuristic

    return TSNE(
        n_components=2,
        perplexity=perplexity,
        learning_rate="auto",
        init="random",
        random_state=42
    ).fit_transform(X)



def run_umap(X):
    X = np.array(X)  # <-- FIX
    return umap.UMAP(
        n_components=2,
        n_neighbors=10,
        min_dist=0.1,
        random_state=42
    ).fit_transform(X)

```

```python
X_a, y_a, l_a = collect_vectors_matrix(animals, pt_embeddings, pt_word2idx, 0)
X_f, y_f, l_f = collect_vectors_matrix(foods, pt_embeddings, pt_word2idx, 1)

X = X_a + X_f
y = y_a + y_f
labels = l_a + l_f

X_tsne = run_tsne(X)
plot_2d(X_tsne, y, labels, "PyTorch Word2Vec — t-SNE")

X_umap = run_umap(X)
plot_2d(X_umap, y, labels, "PyTorch Word2Vec — UMAP")
```

```python
X_a, y_a, l_a = collect_vectors_matrix(animals, gs_embeddings, gs_word2idx, 0)
X_f, y_f, l_f = collect_vectors_matrix(foods, gs_embeddings, gs_word2idx, 1)

X = X_a + X_f
y = y_a + y_f
labels = l_a + l_f

X_tsne = run_tsne(X)
plot_2d(X_tsne, y, labels, "Gensim Word2Vec — t-SNE")

X_umap = run_umap(X)
plot_2d(X_umap, y, labels, "Gensim Word2Vec — UMAP")
```

```python
import gensim
gn = gensim.models.keyedvectors.KeyedVectors.load(
    "word2vec-google-news-300.model",
    mmap="r"
)
X_a, y_a, l_a = collect_vectors_kv(animals, gn, 0)
X_f, y_f, l_f = collect_vectors_kv(foods, gn, 1)

X = X_a + X_f
y = y_a + y_f
labels = l_a + l_f

X_tsne = run_tsne(X)
plot_2d(X_tsne, y, labels, "Google News Word2Vec — t-SNE")

X_umap = run_umap(X)
plot_2d(X_umap, y, labels, "Google News Word2Vec — UMAP")
```

```python
ft = gensim.models.keyedvectors.KeyedVectors.load(
    "fasttext-wiki-news-subwords-300.model",
    mmap="r"
)

X_a, y_a, l_a = collect_vectors_kv(animals, ft, 0)
X_f, y_f, l_f = collect_vectors_kv(foods, ft, 1)

X = X_a + X_f
y = y_a + y_f
labels = l_a + l_f

X_tsne = run_tsne(X)
plot_2d(X_tsne, y, labels, "FastText — t-SNE")

X_umap = run_umap(X)
plot_2d(X_umap, y, labels, "FastText — UMAP")
```

## Short Answer Questions

### Which embeddings performed best at capturing sentence meaning?

The pretrained embeddings (especially Google News Word2Vec and FastText) performed best at capturing sentence-level meaning.

This was evident because:

- They assigned high similarity scores to sentences that were semantically related to the query even when there was little or no word overlap (e.g., cat ↔ feline, sleeping ↔ resting, carpet ↔ rug).

- In the visualization experiments, words belonging to the same semantic category (animals vs foods) formed tight, well-separated clusters, indicating strong semantic structure in the embedding space.

The Gensim-trained Word2Vec performed moderately well:

- It captured some synonym relationships (e.g., cat ↔ feline) better than the PyTorch model.

- However, clusters were less compact and sometimes overlapped.

The PyTorch-trained Word2Vec performed the weakest:

- Sentence similarity was driven mostly by direct word overlap (e.g., presence of cat).

- Synonym-based sentences received low similarity scores.

- In visualizations, animals and foods showed weaker separation and more overlap, indicating less robust semantic representations.


### Were the pretrained embeddings significantly better than your own? Why or why not?
Yes, the pretrained embeddings were significantly better than the embeddings trained from scratch.

Reasons:

- Pretrained models were trained on massive corpora (billions of tokens), allowing them to learn rich semantic relationships and rare word associations.

- They captured synonymy, relatedness, and category structure much more effectively than models trained on a smaller dataset.

- In contrast, the PyTorch and Gensim models were trained on a limited corpus, which restricted their ability to learn nuanced semantic patterns.

Even though the PyTorch implementation was correct, data scale, not algorithmic correctness, was the main limiting factor.


### What factors might explain the differences in performance?

Several factors explain the observed differences:

(a) **Training data size and diversity**

Pretrained embeddings benefit from orders of magnitude more data than the local training corpus.

Larger datasets expose models to more contexts, synonyms, and linguistic variations.

(b) **Vocabulary coverage**

Pretrained embeddings include well-trained vectors for rare or abstract words (e.g., feline, canine), while locally trained models may have weak or noisy representations for such words.

(c) **Training maturity**

Pretrained models are trained for many epochs with carefully tuned hyperparameters, while the local models were trained for a fixed, relatively small number of epochs.

(d) **Sentence representation method**

Sentence embeddings were computed using mean pooling, which treats all words equally.

This can cause common or stop words to dominate the sentence representation, especially in weaker embeddings.

Stronger embeddings are more robust to this limitation.

(e) **Model architecture and enhancements**

FastText benefits from subword information, allowing it to represent morphologically related or rare words better than standard Word2Vec.

```python
#Bonus
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def most_similar(vec, embeddings, idx2word, topk=5):
    sims = cosine_similarity(vec.reshape(1, -1), embeddings)[0]
    top_idx = np.argsort(sims)[::-1][:topk]
    return [(idx2word[i], sims[i]) for i in top_idx]
def analogy(a, b, c, embeddings, word2idx, idx2word, topk=5):
    if a not in word2idx or b not in word2idx or c not in word2idx:
        return "One or more words not in vocabulary."

    va = embeddings[word2idx[a]]
    vb = embeddings[word2idx[b]]
    vc = embeddings[word2idx[c]]

    target = vb - va + vc
    return most_similar(target, embeddings, idx2word, topk)
def analogy_kv(a, b, c, kv, topk=5):
    if not all(w in kv for w in [a, b, c]):
        return "One or more words not in vocabulary."
    return kv.most_similar(positive=[b, c], negative=[a], topn=topk)
```

```python
with open("word2vec_embeddings.pkl", "rb") as f:
    pt = pickle.load(f)

pt_embeddings = np.asarray(pt["embeddings"])
pt_word2idx = pt["word2idx"]
pt_i2w = pt["idx2word"]   # <-- THIS LINE WAS MISSING

analogy("man", "king", "woman", pt_embeddings, pt_word2idx, pt_i2w)
```
