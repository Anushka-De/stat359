# Error Type Analysis

```python
import os, glob, json
import pandas as pd
import matplotlib.pyplot as plt

BASE_DIR = "arithmetic_llm/evaluation_results"
folders = {
    "1–20":  "q1_range_1_20",
    "1–99":  "q1_range_1_99",
    "1–999": "q1_range_1_999",
}

def load_latest_samples(folder):
    files = sorted(glob.glob(os.path.join(BASE_DIR, folder, "sample_outputs_*.json")))
    if not files:
        raise FileNotFoundError(f"No sample_outputs_*.json in {folder}")
    path = files[-1]
    with open(path, "r") as f:
        samples = json.load(f)
    return samples, path

def categorize_sample(s):
    text = s.get("generated_text", "") or ""
    pred = s.get("predicted", None)
    gt   = s.get("ground_truth", None)

    # If predicted is missing, it's non-parseable
    if pred is None:
        # diagnose format error type
        has_think_open = "<think>" in text
        has_think_close = "</think>" in text
        has_final = "Final Result" in text or "Final Result:" in text or "Final Result :" in text

        if has_think_open and not has_think_close:
            return "non_parseable_unclosed_think"
        if not has_final:
            return "non_parseable_missing_final_result"
        return "non_parseable_other"

    # Parseable
    if pred == gt:
        return "correct"
    return "wrong_answer"
```

```python
rows = []
for rng, folder in folders.items():
    samples, path = load_latest_samples(folder)
    for s in samples:
        rows.append({"Range": rng, "Category": categorize_sample(s)})

df = pd.DataFrame(rows)
summary = df.groupby(["Range", "Category"]).size().reset_index(name="Count")
summary["Percent"] = summary.groupby("Range")["Count"].transform(lambda x: 100*x/x.sum())
summary
```

```python
order = ["correct", "wrong_answer", "non_parseable_unclosed_think"]

pivot = summary.pivot(index="Range", columns="Category", values="Percent").fillna(0)
pivot = pivot[[c for c in order if c in pivot.columns]]

ax = pivot.plot(kind="bar", stacked=True, figsize=(8,6))

plt.ylabel("Percent of samples")
plt.title("Error Breakdown by Operand Range")

# Add percentage labels inside bars
for i, row in enumerate(pivot.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 2:  # avoid clutter for very small slices
            ax.text(
                i,
                cumulative + val/2,
                f"{val:.1f}%",
                ha="center",
                va="center",
                fontsize=10,
                color="white",
                weight="bold"
            )
        cumulative += val

plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

- The largest category of errors across all ranges is **wrong_answer**. This shows that the model still generates well-formed outputs, but the computed results are incorrect. Thus, the failure is primarily due to incorrect arithmetic reasoning rather than formatting issues.

- For the largest range (1–999), around 20% of outputs contain an unclosed <think> block, meaning the reasoning process terminates improperly.
This suggests that when expressions become more complex, the model's reasoning chain becomes unstable.

- Interestingly, no unclosed <think> errors appear for the 1–99 range. This is likely due to sampling variability in the evaluation set rather than a systematic improvement in reasoning stability. When evaluated on larger operands (1–999), incomplete reasoning traces reappear, suggesting that longer arithmetic chains increase the probability of generation truncation.

## Discussion

To evaluate the model’s ability to generalize beyond its training distribution, we tested it on arithmetic expressions with increasing operand ranges: 1–20, 1–99, and 1–999. The results show a clear degradation in performance as the numbers become larger.

When evaluated on expressions within the training-like range (1–20), the model achieves approximately 50% accuracy, indicating that it has learned to solve a substantial portion of the arithmetic problems seen during training. However, when the operand range increases to 1–99, accuracy drops to roughly 30%, and for 1–999, correct predictions almost disappear.

Error analysis further reveals that most failures are wrong answers rather than formatting errors. Even when the model produces incorrect results, it typically generates outputs that follow the expected reasoning structure with <think> blocks and a final answer. This indicates that the model has learned the output format and reasoning style, but not the underlying arithmetic rules required for accurate computation.

Interestingly, formatting errors such as unclosed <think> tags appear more frequently for larger operand ranges. This suggests that as arithmetic expressions become more complex, the model’s reasoning chain becomes less stable and more prone to incomplete generation.

Overall, these results suggest that the model **primarily learns patterns within the training distribution rather than general arithmetic reasoning**. While it can replicate the structure of step-by-step solutions, its ability to correctly compute results deteriorates rapidly when faced with numbers outside the range seen during training.

```python
import os, glob, json
import numpy as np
import matplotlib.pyplot as plt

BASE = "arithmetic_llm/evaluation_results"

depths = [4, 6, 8]
ranges = [(1,20), (1,99), (1,999)]
range_labels = ["1–20", "1–99", "1–999"]

def latest_metrics(folder):
    files = sorted(glob.glob(os.path.join(folder, "evaluation_metrics_*.json")))
    if not files:
        return None
    with open(files[-1], "r") as f:
        return json.load(f)

# Accuracy grid (rows=depth, cols=range)
acc = np.full((len(depths), len(ranges)), np.nan)

for i, d in enumerate(depths):
    for j, (lo, hi) in enumerate(ranges):
        folder = os.path.join(BASE, f"grid_depth{d}_range{lo}_{hi}")
        m = latest_metrics(folder)
        if m is None:
            print("Missing metrics in:", folder)
            continue
        # Your metrics are already in percent units (e.g., 43.9), NOT 0-1
        acc[i, j] = m.get("exact_match_accuracy")

# Plot heatmap
plt.figure(figsize=(7,5))
im = plt.imshow(acc, aspect="auto")

plt.xticks(range(len(range_labels)), range_labels)
plt.yticks(range(len(depths)), [str(d) for d in depths])
plt.xlabel("Operand range")
plt.ylabel("Expression depth")
plt.title("Accuracy Heatmap Across Range × Depth")
plt.colorbar(im, label="Exact match accuracy (%)")

# Annotate each cell with the percent
for i in range(acc.shape[0]):
    for j in range(acc.shape[1]):
        if np.isfinite(acc[i, j]):
            plt.text(j, i, f"{acc[i,j]:.1f}%", ha="center", va="center", fontsize=10)

plt.tight_layout()
plt.show()
```

The heatmap shows that numerical range is the dominant factor affecting generalization, while reasoning depth has a comparatively smaller effect. This suggests that the model primarily struggles with arithmetic computation involving larger numbers, rather than maintaining multi-step reasoning.

These results support the hypothesis that the model learns patterns within the training distribution rather than a generalizable algorithm for arithmetic.

# Error Magnitude

```python
import os
import glob
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

BASE = "arithmetic_llm/evaluation_results"

depths = [4, 6, 8]
ranges = [(1, 20), (1, 99), (1, 999)]

def latest_samples(folder):
    files = sorted(glob.glob(os.path.join(folder, "sample_outputs_*.json")))
    if not files:
        return None
    with open(files[-1], "r") as f:
        return json.load(f)

rows = []

for d in depths:
    for lo, hi in ranges:
        folder = f"{BASE}/grid_depth{d}_range{lo}_{hi}"
        samples = latest_samples(folder)
        if samples is None:
            print("Missing samples:", folder)
            continue

        for s in samples:
            pred = s.get("predicted")
            gt = s.get("ground_truth")

            if gt is None:
                continue

            # Extract an integer from pred even if pred is messy (string etc.)
            match = re.search(r"-?\d+", str(pred))
            if not match:
                continue

            try:
                pred_int = int(match.group())
                gt_int = int(gt)
            except:
                continue

            error = abs(pred_int - gt_int)

            rows.append({
                "Range": f"{lo}-{hi}",
                "Depth": f"d={d}",
                "Error": error
            })

df_error = pd.DataFrame(rows)
df_error
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df_error,
    x="Range",
    y="Error",
    hue="Depth",
    palette="Set2"
)

plt.yscale("log")  # important because errors grow very large

plt.xlabel("Operand Range")
plt.ylabel("Absolute Error (log scale)")
plt.title("Error Magnitude vs Operand Range and Expression Depth")

plt.legend(title="Expression Depth")
plt.tight_layout()

plt.show()
```

-  **Error magnitude increases dramatically with operand range**

The most striking pattern is the rapid increase in error magnitude as the operand range grows. Error magnitude increases dramatically with operand range

- **Operand range has a stronger effect than reasoning depth**

While expression depth increases from 4 → 6 → 8, the error distributions remain relatively similar within each range. While expression depth increases from 4 → 6 → 8, the error distributions remain relatively similar within each range.

- **Error variability increases under OOD conditions**

As the operand range increases, the spread of the error distribution becomes much wider, indicating more unstable predictions.

- **Outliers indicate severe reasoning failures**

The plot shows several extreme outliers (especially for 1–999), where errors reach thousands.

These cases occur when the model produces reasoning steps that lead to completely incorrect arithmetic results.

Key Conclusion: The arithmetic LLM struggles primarily with numerical generalization, rather than with increasing reasoning depth. The arithmetic LLM struggles primarily with numerical generalization, rather than with increasing reasoning depth.
