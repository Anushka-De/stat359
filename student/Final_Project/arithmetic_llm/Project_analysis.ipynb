# Error Type Analysis

```python
import os, glob, json
import pandas as pd
import matplotlib.pyplot as plt

BASE_DIR = "arithmetic_llm/evaluation_results"
folders = {
    "1–20":  "q1_range_1_20",
    "1–99":  "q1_range_1_99",
    "1–999": "q1_range_1_999",
}

def load_latest_samples(folder):
    files = sorted(glob.glob(os.path.join(BASE_DIR, folder, "sample_outputs_*.json")))
    if not files:
        raise FileNotFoundError(f"No sample_outputs_*.json in {folder}")
    path = files[-1]
    with open(path, "r") as f:
        samples = json.load(f)
    return samples, path

def categorize_sample(s):
    text = s.get("generated_text", "") or ""
    pred = s.get("predicted", None)
    gt   = s.get("ground_truth", None)

    # If predicted is missing, it's non-parseable
    if pred is None:
        # diagnose format error type
        has_think_open = "<think>" in text
        has_think_close = "</think>" in text
        has_final = "Final Result" in text or "Final Result:" in text or "Final Result :" in text

        if has_think_open and not has_think_close:
            return "non_parseable_unclosed_think"
        if not has_final:
            return "non_parseable_missing_final_result"
        return "non_parseable_other"

    # Parseable
    if pred == gt:
        return "correct"
    return "wrong_answer"
```

```python
rows = []
for rng, folder in folders.items():
    samples, path = load_latest_samples(folder)
    for s in samples:
        rows.append({"Range": rng, "Category": categorize_sample(s)})

df = pd.DataFrame(rows)
summary = df.groupby(["Range", "Category"]).size().reset_index(name="Count")
summary["Percent"] = summary.groupby("Range")["Count"].transform(lambda x: 100*x/x.sum())
summary
```

```python
order = ["correct", "wrong_answer", "non_parseable_unclosed_think"]

pivot = summary.pivot(index="Range", columns="Category", values="Percent").fillna(0)
pivot = pivot[[c for c in order if c in pivot.columns]]

ax = pivot.plot(kind="bar", stacked=True, figsize=(8,6))

plt.ylabel("Percent of samples")
plt.title("Error Breakdown by Operand Range")

# Add percentage labels inside bars
for i, row in enumerate(pivot.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 2:  # avoid clutter for very small slices
            ax.text(
                i,
                cumulative + val/2,
                f"{val:.1f}%",
                ha="center",
                va="center",
                fontsize=10,
                color="white",
                weight="bold"
            )
        cumulative += val

plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

- The largest category of errors across all ranges is **wrong_answer**. This shows that the model still generates well-formed outputs, but the computed results are incorrect. Thus, the failure is primarily due to incorrect arithmetic reasoning rather than formatting issues.

- For the largest range (1–999), around 20% of outputs contain an unclosed <think> block, meaning the reasoning process terminates improperly.
This suggests that when expressions become more complex, the model's reasoning chain becomes unstable.

- Interestingly, no unclosed <think> errors appear for the 1–99 range. This is likely due to sampling variability in the evaluation set rather than a systematic improvement in reasoning stability. When evaluated on larger operands (1–999), incomplete reasoning traces reappear, suggesting that longer arithmetic chains increase the probability of generation truncation.

## Discussion

To evaluate the model’s ability to generalize beyond its training distribution, we tested it on arithmetic expressions with increasing operand ranges: 1–20, 1–99, and 1–999. The results show a clear degradation in performance as the numbers become larger.

When evaluated on expressions within the training-like range (1–20), the model achieves approximately 50% accuracy, indicating that it has learned to solve a substantial portion of the arithmetic problems seen during training. However, when the operand range increases to 1–99, accuracy drops to roughly 30%, and for 1–999, correct predictions almost disappear.

Error analysis further reveals that most failures are wrong answers rather than formatting errors. Even when the model produces incorrect results, it typically generates outputs that follow the expected reasoning structure with <think> blocks and a final answer. This indicates that the model has learned the output format and reasoning style, but not the underlying arithmetic rules required for accurate computation.

Interestingly, formatting errors such as unclosed <think> tags appear more frequently for larger operand ranges. This suggests that as arithmetic expressions become more complex, the model’s reasoning chain becomes less stable and more prone to incomplete generation.

Overall, these results suggest that the model **primarily learns patterns within the training distribution rather than general arithmetic reasoning**. While it can replicate the structure of step-by-step solutions, its ability to correctly compute results deteriorates rapidly when faced with numbers outside the range seen during training.

```python
#Plotting the Accuracy
import pandas as pd
import matplotlib.pyplot as plt
depth = [4,6,8]
accuracy = [43.9, 43.0, 41.6]

plt.figure(figsize=(6,4))
plt.plot(depth, accuracy, marker="o", linewidth=2)
for x, y in zip(depth, accuracy):
    plt.text(x + 0.1, y, f"{y:.1f}%", fontsize=10, va="center")
plt.xlabel("Expression depth")
plt.ylabel("Exact match accuracy (%)")
plt.title("Accuracy vs Expression Depth")
plt.show()
```

## Interpretation: Accuracy vs Expression Depth

The plot shows how model accuracy changes as the expression depth increases while keeping the operand range fixed at 1–20.

The results indicate that increasing the expression depth leads to only a small decrease in accuracy. When evaluated on expressions with depth 4, the model achieves an accuracy of approximately 43.9%. Increasing the depth to 6 reduces accuracy slightly to 43.0%, and further increasing the depth to 8 results in an accuracy of about 41.6%.

This relatively small drop suggests that the model is capable of handling moderately longer reasoning chains when the arithmetic operands remain within the training range. In other words, the model does not struggle significantly with deeper expression structures as long as the numerical values remain simple.

This result contrasts with the earlier experiment on larger operand ranges, where accuracy dropped dramatically. Therefore, the findings suggest that the primary limitation of the model lies in generalizing to larger numbers rather than handling deeper reasoning structures.
